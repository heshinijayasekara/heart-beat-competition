```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
train_values = pd.read_csv("train_values.csv")
train_values.head()
```

```python
train_labels = pd.read_csv("train_labels.csv")
train_labels.head()
```

```python
df = train_values.merge(train_labels,on="patient_id")
df.head()
```

```python
positive_df = df.loc[df.heart_disease_present == 1,:]
negative_df = df.loc[df.heart_disease_present == 0,:]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), sharey=True, dpi=120)
ax1.bar(negative_df.thal.unique(),negative_df.thal.value_counts())
ax2.bar(positive_df.thal.unique(),positive_df.thal.value_counts())
plt.xlabel("thal")
plt.show()
```
